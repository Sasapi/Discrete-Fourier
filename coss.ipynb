```python
def coss(m):
    %matplotlib inline
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.arange(0, m, 0.1)
    y = np.cos(x)
    plt.plot(x, y)
    plt.show()
    
coss(20)
```

```python
def coss(p):
    %matplotlib inline
    import matplotlib.pyplot as plt
    import numpy as np
    from numpy import pi
    x = np.linspace(0, 2*pi, p)
    y = np.cos(x)
    plt.plot(x, y)
    plt.show()
    print(y)

coss(100)
```
